# Abstract

Pair WMO Codes in (2) with corresponding Weather Descriptions accessed in (3).

```python
import pandas as pd
```

```python
%store -r weather
```

```python
%store -r wmo
```

# Pair Weather Code + Weather Description

```python
weather = pd.merge(weather, wmo, on='Weather Code (WMO)', how='left')
```

```python
weather
```

# Editing Data Frame

### Rename Columns

```python
newColNames = {
    'Weather Descrption': 'Weather Description'
}
```

```python
weather.rename(columns = newColNames, inplace = True)
```

### Reorder Columns

```python
weather = weather[['Date', 'Latitude', 'Longitude', 'Avg. Apparent Temperature (F)', 'Avg. Temperature at 2m (F)',
                   'Sunshine Duration (sec)', 'Total Shortwave Radiation (MJ/m2)', 
                  'Total Precipitation (mm)', 'Total Precipitation Time (hrs)', 'Total Shortwave Radiation (MJ/m2)',
                  'Evapotranspiration (mm)', 'Weather Code (WMO)', 'Weather Description']]
```

### Remove Duplicate Rows

```python
weather = weather.loc[:, ~weather.columns.duplicated()]
```
