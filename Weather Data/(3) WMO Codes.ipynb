# Editing WMO Dictionary

### Remove 'night' Keys

```python
filtered_wmo_dict = {key: {subkey.replace('night', ''): value for subkey, value in subdict.items()} for key, subdict in wmo_dict.items()}
```

### Convert to Data Frame

```python
wmo_list = [{'key': key, 'description': subdict['day']['description']} for key, subdict in filtered_wmo_dict.items()]
```

```python
wmo = pd.DataFrame(wmo_list)
```

### Rename Columns

```python
newColNames = {
    'key': 'Weather Code (WMO)',
    'description': 'Weather Descrption'
}
```

```python
wmo.rename(columns = newColNames, inplace = True)
wmo
```

```python
wmo['Weather Code (WMO)'] = wmo['Weather Code (WMO)'].astype('float64')
```
