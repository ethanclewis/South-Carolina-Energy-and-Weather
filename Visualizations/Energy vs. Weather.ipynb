# Abstract

Dashboards to visualize relationships between South Carolina electricty demand / in-state energy generation methods and weather conditions.

```python
%store -r energy_SC
%store -r avg_weather
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

# Clean Energy Generation by Weather Condition (POLAR PLOT)

### Create Data Frame Subset

```python
energy_SC['Date'] = pd.to_datetime(energy_SC['Date'])
avg_weather['Date'] = pd.to_datetime(avg_weather['Date'])

weather_energy = pd.merge(energy_SC, avg_weather, on = ['Date'], how = 'left')
```

```python
weather_energy = weather_energy[weather_energy['Measurement Code'].isin(['NUC','SUN','WAT','D'])]

weather_energy = weather_energy.groupby(['Measurement', 'Weather Description']).mean().reset_index()
```

```python
weather_energy
```

### Assign Colors and Appearance Order

```python
custom_color_map_1 = {
    'Demand': 'rgb(228, 26, 28)', # Red
    'Nuclear': 'rgb(148, 103, 189)',   # Yellow
    'Hydro': 'rgb(77, 175, 74)',  # Green
    'Solar': 'rgb(255, 217, 47)' # Purple
}


custom_order_1 = ['Demand', 'Nuclear', 'Hydro', 'Solar']
custom_colors_1 = [custom_color_map_1[measurement] for measurement in custom_order_1]
```

### Generate Plot

```python
weather_energy_polar = px.line_polar(weather_energy, r="MWh", theta="Weather Description", color = 'Measurement', 
                    line_close=True, color_discrete_sequence=custom_colors_1, 
                    title = "South Caroline Energy Demand and Clean Energy Generation by Weather Conditions")

weather_energy_polar.update_layout(
    legend=dict(x=0.9),  
)

weather_energy_polar.update_traces(
    line=dict(width=3)  
)

weather_energy_polar.show()
```

# Energy DEMAND by Weather Condition

### List of Available Weather Measurements

```python
avg_weather.columns
```

### Create Weather Plot (Use List Above to Customize)

```python
demand = go.Figure()

demand.add_trace(go.Scatter(x=avg_weather['Date'], 
                         y=avg_weather['Avg. Apparent Temperature (F)'], #CHANGE WEATH MEASUREMENT
                         
                         mode='markers', #CHANGE PLOT TYPE
                         
                         name='Average Temperature', #CHANGE LABEL
                         
                         line=dict(color= 'rgb(55, 126, 184)'))) #CHANGE COLOR
pass
```

### Create Demand Plot (DO NOT CHANGE)

```python
demand.add_trace(go.Scatter(x=energy_SC[energy_SC['Measurement'] == 'Demand']['Date'], 
                         y=energy_SC[energy_SC['Measurement'] == 'Demand']['MWh'], mode='lines', 
                         name= 'Demand', 
                         line=dict(color = '#E45756'), yaxis='y2'))
pass
```

### Generate Plot (Customize Based on Weather Measurement)

```python
demand.update_layout(
    title='South Carolina - Electricity Demand and Average Temperature', #CHANGE TITLE
    
    xaxis=dict(title='Date'),
    
    yaxis=dict(title=f'Temperature (\u00b0F)', side='left', showgrid=False,overlaying='y2',), #CHANGE AXIS
    
    yaxis2=dict(title='MWh', showgrid=False,side='right'),
    legend=dict(x=1.1, title = 'Measurement')
)

demand.show()
```
